```python
def norm_func(i):
    x = (i-i.min())	/	(i.max()	-	i.min())
    return (x)
```

```python
spotify.info()
```

```python
spotify.drop("artist", axis=1,inplace=True)
```

```python
spotify.drop('song', axis=1,inplace=True)
```

```python
spotify['genre'].value_counts()
```

```python
spotify.drop('genre', axis=1,inplace=True)
```

```python
spotify.drop('explicit', axis=1,inplace=True)
```

```python
spotify_norm = norm_func(spotify.iloc[:,1:])
```

```python
spotify_norm.head(10)
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    clf.fit(spotify_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS

plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10,  random_state=0)
y_kmeans = clf.fit_predict(spotify_norm)
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
md=pd.Series(y_kmeans)
spotify['clust']=md
spotify
```
